```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get('https://www.webscraper.io/test-sites/e-commerce/static')
```

```python
soup = BeautifulSoup(r.content, 'html.parser') 
print(soup.prettify()) 
```

```python
divs = soup.find_all('div', class_= "card-body")
```

```python
len(divs)
```

```python
table_data = []

for card in divs:
    image = card.find('img')['src']
    price = card.find('h4', class_='price').text
    title = card.find('a', class_='title').text
    description = card.find('p', class_='description').text
    review_count = card.find('p', class_='review-count').text
    rating = len(card.find_all('span', class_='ws-icon-star'))
    
    table_data.append([image, title, price, description, review_count, rating])
```

```python
pd.DataFrame(table_data,columns=['Image_Link', 'Name','Price','Features','N_reviews','Stars'])
```
