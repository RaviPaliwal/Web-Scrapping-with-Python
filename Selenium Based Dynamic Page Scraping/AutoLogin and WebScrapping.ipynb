```python
# Importing Required Libraries
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
# Configure the webdriver for Edge Browser
driver = webdriver.Edge()
```

```python
# Opening The Webpage
url = "https://indianexpress.com/login/"
driver.get(url)
```

```python
# Login function
def login(email,password):
    email_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "textNumberInput")))
    # Clear previous Inputs enter Email
    email_input.clear()
    email_input.send_keys(email)

    # Continue
    continue_btn = WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.ID, "multiAuthSubmit")))
    continue_btn.click()

    # Password Filling
    pass_input = WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.ID, "horizontal_password")))
    pass_input.send_keys(password)

    # Signing in
    sign_in_btn = WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.ID, "loginRecaptcha")))
    sign_in_btn.click()
```

### I am targetting sports news articles data so switching to sports tab and collecting title date content and image link data.

```python

# Switching to Sports Tab
SportsTab = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, "//a[contains(@class, 'ie_ga_event_tracking') and contains(text(), 'Sports')]")))
SportsTab.click()
# Switching Driver to new Window Opened
driver.switch_to.window(driver.window_handles[1])
```

### Extracting data Dynamicaly Clicking on Next Page. Close Tab To Stop

```python
titles = []
dates = []
contents = []
images = []

while True:
    # Delay to load page Properly
    time.sleep(5)
    try:
        # Getting Next Button
        next_btn = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CLASS_NAME, "next")))
        # Getting Html of page
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # find all Articles
        articles = soup.find_all(class_='articles')
        for article in articles:
            title = article.find("h2", class_="title").text.strip()
            date = article.find("div", class_="date").text.strip()
            content = article.find("div", class_="img-context").text.strip()
            image_url = article.find("div", class_="snaps").find("img")["src"]

            titles.append(title)
            dates.append(date)
            contents.append(content)
            images.append(image_url)

        # Clicking on Next
        next_btn.click()

    except Exception as e:
        print(f"Either Interrupted or Some Error")
        break
```

```python
data = {
    "Title": titles,
    "Date": dates,
    "Content": contents,
    "Image_URL": images
}
df = pd.DataFrame(data)

df
```
