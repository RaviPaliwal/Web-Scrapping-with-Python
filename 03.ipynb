```python
Json = []
progress_bar = tqdm(movies_list, desc="Processing")
for title in progress_bar:
#     for title in movies_list

    url = f"https://www.omdbapi.com/?s={title}&apikey=ef22cdcf"

    # Send a GET request to the OMDB API
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the JSON data from the response
        data = response.json()

        # Check if the 'Response' key is present and its value is not "False"
        if 'Response' in data and data['Response'] == "False":
            continue
        else:
            # Append the 'Search' results to the Json list
            Json.append(data['Search'])
    else:
        # Print an error message if the request was not successful
        print(f"Error: Unable to fetch data for {title}")
```

```python
Json[0][1]
```

```python
Data =[]
for sub_array in Json:
    for i in sub_array:
        obj = {
            'Title': i['Title'],
            'Year': i['Year'],
            'Type': i['Type'],
            'Poster': i['Poster']
        }
    # Append the object to the list
    Data.append(obj)
```
